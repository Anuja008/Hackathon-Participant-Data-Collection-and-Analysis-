```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("HackData.csv")

# Fill numeric missing values with mean
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())

# Fill categorical missing values with mode
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Verify again
print("Missing after fix:", df.isnull().sum().sum())
```

```python
# Percentage of missing values
print("\nPercentage of missing values:\n")
print((df.isnull().sum() / len(df)) * 100)
```

```python
# Step 2 — Handle Missing Values (fix remaining 720)

# Fill numeric missing values with mean (skip fully-null columns)
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    if df[col].notnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mean())

# Fill categorical missing values with mode (skip fully-null columns)
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    if df[col].notnull().sum() > 0 and not df[col].mode().empty:
        df[col] = df[col].fillna(df[col].mode()[0])

# Drop columns that are completely null
df = df.dropna(axis=1, how='all')

# Final verification
print("Missing after fix:", df.isnull().sum().sum())
```

```python
# Step 3 — Remove Duplicate Records

# Identify duplicates
print("Duplicates found:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# Verify
print("Rows after removing duplicates:", len(df))
```

```python
# Step 4 — Detect and Handle Outliers (IQR Method)

numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print("Rows after outlier removal:", len(df))
```

```python
# Save clean dataset
df.to_csv("cleaned_HackData.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# The original error was NotFittedError because the model was not trained.
# Additionally, X_train was empty from previous steps.
# We need to re-establish the features, target, preprocessing, and model training.

# Reload the cleaned dataset to ensure consistent state
df = pd.read_csv("cleaned_HackData.csv")

# Define the target variable
y = df['Team Members']

# Select categorical features for X (as 'Team Members' was the only numeric one)
categorical_features = ['College Name', 'Track Chosen']
X = df[categorical_features]

# Create a preprocessor for one-hot encoding
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough' # In this case, there are no other columns to pass through
)

# Apply preprocessing to the features
X_processed = preprocessor.fit_transform(X)

# Get feature names after one-hot encoding for the DataFrame creation
feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
X_processed_df = pd.DataFrame(X_processed.toarray(), columns=feature_names, index=X.index)

# Step 3: Split Data into Training and Testing Sets (80:20)
X_train, X_test, y_train, y_test = train_test_split(
    X_processed_df, y, test_size=0.2, random_state=42
)

print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))

# Step 4: Train the Linear Regression Model (This was missing or removed)
model = LinearRegression()
model.fit(X_train, y_train)

print("Intercept:", round(model.intercept_, 2))

coeff_df = pd.DataFrame({
    "Feature": X_train.columns, # Use X_train.columns as it now contains encoded feature names
    "Coefficient": model.coef_
})
print(coeff_df)

# Step 5: Evaluate the Model (Original content of the cell)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"\nMSE : {mse:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R2 Score: {r2:.4f}")
```

```python
# Step 6: Sample Prediction

# Use one sample from test data
sample = X_test.iloc[0].values.reshape(1, -1)

predicted_value = model.predict(sample)

print(f"Predicted Value: {predicted_value[0]:.2f}")
```
